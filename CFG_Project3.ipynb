```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path= '/content/drive/MyDrive/CFG python project 3/sales.csv' 
```

```python

def read_data():
  data = []
  with open(data_path, 'r') as sales_csv:
    spreadsheet = csv.DictReader(sales_csv)
    for row in spreadsheet:
      data.append(row)
  return data

def run():
  data = read_data()
  sales = []
  for row in data:
    sale = int(row['sales'])
    sales.append(sale)
  total = sum(sales)
  print('Total sales: {}'.format(total))
  return sales


month_sales= run()
```

```python
month_sales
```

```python
fig , ax = plt.subplots(1,1,figsize=(12,12))
ax.plot(month_sales)
plt.show()
```

```python
def sales_summary(data):
  percent_change =[]
  avg_sales=[]
  sales_high=[]
  sales_low=[]
  for i,j in enumerate(data):
    if i==0:
      percent_change.append(round(0,2))
    else:
      sales_change= ((data[i]-data[i-1])/data[i-1])*100
      percent_change.append(round(sales_change,2))
  avg_sales= [round(sum(data)/len(data),2)]*len(data)
  sales_high=[round(max(data),2)]*len(data)
  sales_low=[round(min(data),2)]*len(data)
  summary_df = pd.DataFrame({
      'Month': ['Jan','Feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec'],
      'Monthly Sales': month_sales,
      'Perecnt Change': percent_change,
      'Avg Sales': avg_sales,
      'Max Sales': sales_high,
      'Min Sales': sales_low,
  })
  return summary_df

```

```python
summary_df = sales_summary(month_sales)
```
